## Outliers

```python
import seaborn as sns
# Visualize outliers using a box plot
sns.boxplot(x=df['instant'])
```

```python
# Identify outliers using z-scores
from scipy import stats

z_scores = stats.zscore(df['temp'])
outliers = (z_scores > 3) | (z_scores < -3)
outlier_values = df['temp'][outliers]
print(outlier_values)
```

```python
q1 = new_dataset['windspeed'].quantile(.25)
q2 = new_dataset['windspeed'].quantile(.50)
q3 = new_dataset['windspeed'].quantile(.75)
```

```python
def find_outliers(df, cols):
    q1 = df[cols].quantile(.25)
    q3 = df[cols].quantile(.75)
    iqr = q3 - q1
    min_r = q1 - 1.5*iqr
    max_r = q3 + 1.5*iqr
    ouliers_indices = df.index[(df[cols] < min_r) | (df[cols] > max_r)]
    return ouliers_indices
```
